```python
import pandas as pd
from surprise import Reader, Dataset
from surprise import SVD, evaluate
from surprise.model_selection import train_test_split
from surprise.dump import dump
from surprise.dump import load
```

```python
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')
ratings = ratings.drop(['timestamp'],axis = 1) #dropping the timestamp variable..not important for this study
```

```python
ratings
```

```python
ratings_dict = {'itemID': list(ratings.loja_id),
                'userID': list(ratings.user_id),
                'rating': list(ratings.like)}
df = pd.DataFrame(ratings_dict)

# A reader is still needed but only the rating_scale param is required.
# The Reader class is used to parse a file containing ratings.
reader = Reader(rating_scale=(0.0, 1.0))

# The columns must correspond to user id, item id and ratings (in that order).
data = Dataset.load_from_df(df[['userID', 'itemID', 'rating']], reader)
```

```python
model = SVD()
evaluate(model, data, measures=['RMSE'])
```

```python
trainset, testset = train_test_split(data, test_size=.25)
model.fit(trainset)
```

```python
testset[0]
```

```python
predictions = model.test(testset)
```

```python
predictions
```

```python
dump('recomendacao_filmes',algo=model)
```
