Criando base artificial de usuários com 3 perfis diferentes.

```python
loja_id = []
user_id = []
like = []
```

```python
prefs = [('VESTUÁRIO', 'ARTIGOS DO LAR'),
         ('VESTUÁRIO', 'ARTIGOS DO LAR'),
         ('VESTUÁRIO'),
         ('VESTUÁRIO'),
         ('VESTUÁRIO'),
         ('ARTIGOS DO LAR'),
         ('ARTIGOS DO LAR'),
         ('VESTUÁRIO', 'ARTIGOS DO LAR'),
         ('VESTUÁRIO', 'ARTIGOS DO LAR'),
         ('VESTUÁRIO'),
         ('VESTUÁRIO'),
         ('VESTUÁRIO'),
         ('ARTIGOS DO LAR'),
         ('ARTIGOS DO LAR'),
         ('VESTUÁRIO', 'ARTIGOS DO LAR'),
         ('VESTUÁRIO', 'ARTIGOS DO LAR'),
         ('VESTUÁRIO'),
         ('VESTUÁRIO'),
         ('VESTUÁRIO'),
         ('ARTIGOS DO LAR'),
         ('ARTIGOS DO LAR')]

for k in range(0, 20):
    for j in range(0, len(prefs)):
        for i in range(0, lojas.shape[0]):
            if random.random() <= 0.5:
                row = lojas.iloc[[0]]
                loja_id.append(i)
                user_id.append(j)
                if lojas.iloc[[0]]['SEGMENTO 2'].values[0] in prefs[j]:
                    if random.random() <= 0.7:
                        like.append(0.5)
                    else:
                        like.append(0)
                else:
                    if random.random() <= 0.3:
                        like.append(0.5)
                    else:
                        like.append(0)
                
```

```python
ratings = pd.DataFrame({'loja_id': loja_id, 'user_id': user_id, 'like': like})
```

```python
ratings
```

```python
ratings.to_csv('ratings.csv', index=False)
```

Montando base artificial de ofertas fornecidas manualmente pelo lojista.

```python
titulo = []
url = []
imagem = []
preco = []
lojas_ids = []
priority = []
for k in range(0, 20):
    for i in range(0, lojas.shape[0]):
        titulo.append("OFERTA " + str(k) + " DA LOJA " + str(i))
        url.append("https://www.instagram.com/p/BmcORVRBlIJ/?taken-at=2121748688067072")
        imagem.append("ursal.png")
        preco.append(random.random()*100)
        lojas_ids.append(i)
        priority.append(random.randint(1,3))
```

```python
ofertas_manuais = pd.DataFrame({'titulo': titulo, 'url': url, 'imagem': imagem, 'preco': preco,
                                'lojas': lojas_ids, 'priority': priority})
```

```python
ofertas_manuais.to_csv('ofertas_manuais.csv', index=False)
```
